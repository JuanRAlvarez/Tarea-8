```python
from scipy.integrate import odeint
alpha=25
beta=1
gamma= 3
delta= 2
def system(z,t):
    x,y=z[0],z[1]
    dxdt= x*(alpha-beta*y)
    dydt=-y*(gamma-delta*x)
    return [dxdt,dydt]
t=linspace(0.06,0.8,1000)
x0,y0=13.0,15.0
sol=odeint(system,[x0,y0],t)
X,Y=sol[:,0],sol[:,1]
plot(t,Y)
plot(t,X)
scatter(datos[:,0],datos[:,1],c='r')
scatter(datos[:,0],datos[:,2])
```
