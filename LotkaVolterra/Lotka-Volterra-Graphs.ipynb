```python
from scipy.integrate import odeint


alpha = 30
beta =  4
sigma = 2
gamma = 50


def LotkaVolterra(state,t, alpha, beta, sigma, gamma):
  x = state[0]
  y = state[1]
  xd = x*(alpha - beta*y)
  yd = -y*(gamma - sigma*x)
  return [xd,yd]

t = linspace(0.0,0.8,500)
state0 = [15,13]
state = odeint(LotkaVolterra,state0,t)
figure()
plot(t,state)
ylim([0,80])
xlim([0,0.8])
xlabel('Time')
ylabel('Population Size')
legend(('x (prey)','y (predator)'))
title('Lotka-Volterra equations')
```

```python
#SOLVING

state0 = [datos[0,1],datos[0,2]]

def LotkaVolterra(state,t, alpha, beta, sigma, gamma):
    x = state[0]
    y = state[1]
    xd = x*(alpha - beta*y)
    yd = -y*(gamma - sigma*x)
    return [xd,yd]

#MODELLING

def my_model(t,alpha,beta, gamma, delta):
    return odeint(LotkaVolterra,state0,t, args=(alpha,beta,gamma, delta))
```
