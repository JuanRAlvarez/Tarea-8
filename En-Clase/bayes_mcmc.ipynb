# Haciendo ajustes usando cadenas de Markov

Tenemos que ajustar un parámetro de movimiento parabólico

```python
%pylab inline
```

```python
datos = loadtxt("movimiento.dat")

x_obs = datos[:,0]
y_obs = datos[:,1]
```

```python
scatter(x_obs,y_obs)
```

```python
def likelihood(y_obs, y_model):
    chi_squared = (1.0/2.0)*sum((y_obs-y_model)**2)
    return -chi_squared
```

```python
def my_model(x_obs, v0, b, x0):
    return v0*x_obs + b * x_obs**2 + x0
```

```python
v0_walk = empty((0)) #this is an empty list to keep all the steps
b_walk = empty((0))
x0_walk = empty((0))
l_walk = empty((0))

v0_walk = append(v0_walk, random.random())
b_walk = append(b_walk, random.random())
x0_walk = append(x0_walk, random.random())

y_init = my_model(x_obs, v0_walk[0], b_walk[0], x0_walk[0])
l_walk = append(l_walk, likelihood(y_obs, y_init))
print v0_walk
print b_walk
print x0_walk
print l_walk
```

```python
n_iterations = 20000 #this is the number of iterations I want to make
for i in range(n_iterations):
    v0_prime = np.random.normal(v0_walk[i], 1.0) 
    b_prime = np.random.normal(b_walk[i], 1.0)
    x0_prime = np.random.normal(x0_walk[i], 1.0)

    y_init = my_model(x_obs, v0_walk[i], b_walk[i], x0_walk[i])
    y_prime = my_model(x_obs, v0_prime, b_prime, x0_prime)
    
    l_prime = likelihood(y_obs, y_prime)
    l_init = likelihood(y_obs, y_init)
    
    alpha = l_prime/l_init
    if(alpha<=1.0):
        v0_walk  = append(v0_walk,v0_prime)
        b_walk  = append(b_walk,b_prime)
        x0_walk = append(x0_walk,x0_prime)
        l_walk = append(l_walk, l_prime)
    else:
        beta = random.random()
        if(beta<=exp(-alpha)):
            v0_walk = append(v0_walk,v0_prime)
            b_walk = append(b_walk,b_prime)
            x0_walk = append(x0_walk, x0_prime)
            l_walk = append(l_walk, l_prime)
        else:
            v0_walk = append(v0_walk,v0_walk[i])
            b_walk = append(b_walk,b_walk[i])
            x0_walk = append(x0_walk, x0_walk[i])
            l_walk = append(l_walk, l_init)
```
